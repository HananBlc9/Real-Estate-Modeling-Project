```python
import pandas as pd
from fastapi import FastAPI, File, UploadFile
import uvicorn
import nest_asyncio
import threading
import io
```

```python
app = FastAPI()

@app.post("/preprocess_data/")
async def preprocess_data(file: UploadFile = File(...)):
    if not file.filename.endswith(".csv"):
        return {"Error": "Only CSV files are allowed!"}
    data = await file.read()

    try:
        df = pd.read_csv(io.BytesIO(data))
    except Exception as e:
        return {"Error": f"Failed to read CSV: {str(e)}"}

    cleanDF =  df.dropna(axis=1, how='all')
    return {"filename": file.filename, "original_rows": len(df), "clean_rows": len(cleanDF)}

nest_asyncio.apply()

def run():
    uvicorn.run(app, host="0.0.0.0",port=8000)

threading.Thread(target=run).start()
```

```python
import json

def extract_amount_max(val):
    try:
        if pd.notnull(val) and val.strip().startswith('['):
            data = json.loads(val)
            if isinstance(data, list) and data and 'amountMax' in data[0]:
                return data[0]['amountMax']
    except (json.JSONDecodeError, TypeError, IndexError, KeyError):
        pass
    return None

df['prices'] = df['prices'].apply(extract_amount_max)
df['prices'] = pd.to_numeric(df['prices'], errors='coerce')
```

```python
merged_df.drop(columns=['roofing','parking'],inplace=True)
# Sample fix function
def clean_parking_types(cell):
    if pd.isna(cell):  # handle NaNs
        return []
    types = [t.strip().lower() for t in cell.split(',')]  # normalize casing and trim
    types = sorted(set(types))  # remove duplicates & sort for consistency
    return ', '.join(types)  # back to string if needed

# Apply cleaning function
merged_df['parkingTypes_cleaned'] = merged_df['parkingTypes'].apply(clean_parking_types)
merged_df.info()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


# ===== 2. Label encode 'lotSizeUnit' =====
le = LabelEncoder()
merged_df['lotSizeUnit_encoded'] = le.fit_transform(merged_df['lotSizeUnit'])

# ===== 3. Define target and features =====
y = merged_df['prices']
X = merged_df.drop('prices', axis=1)

# ===== 4. Split data =====
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ===== 5. Define columns =====
numerical_cols = [
    'floorSizeValue', 'lotSizeValue', 'mostRecentPricePerSquareFoot',
    'numBathroom', 'numBedroom', 'numFloor', 'numParkingSpaces', 'yearBuilt'
]

# Remove 'lotSizeUnit' from categorical columns since it's now label encoded
categorical_cols = [
    'city', 'county', 'mostRecentStatus', 'neighborhoods',
 'parkingTypes_cleaned', 'propertyType', 'province', 'sewerType'
]

# Include multi-label dummy columns if any
multilabel_cols = [col for col in X.columns if col.startswith('exterior_') or col.startswith('hvac_')]

# Combine features, including label-encoded 'lotSizeUnit'
used_features = numerical_cols + categorical_cols + ['lotSizeUnit_encoded'] + multilabel_cols
X_train = X_train[used_features]
X_test = X_test[used_features]

# ===== 6. Preprocessing =====
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
], remainder='passthrough')  # 'lotSizeUnit_encoded' + multi-labels are passed through

# ===== 7. Define models =====
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(max_depth=10,random_state=42),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Lasso Regression': Lasso(alpha=1.0),
    'Ridge Regression': Ridge(alpha=0.1)
}

# ===== 8. Train and evaluate each model =====
for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', model)
    ])

    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} R² Score: {r2:.4f}")
```
